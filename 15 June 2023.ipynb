```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df=pd.read_excel("superstore.xlsx")


```

```python
# Convert 'Quantity' and 'Profit' columns to numeric type
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')

# Convert 'Order Date' column to datetime type
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Calculate the mean of 'Quantity' column
mean_quantity = df['Quantity'].mean()
print("Mean Quantity:", mean_quantity)

# Print selected columns
selected_cols = df[['Order Date', 'Ship Date', 'City']]
print(selected_cols.head())

# Create a line plot
plt.plot(df['Order Date'], df['Profit'])
plt.title('Profit over Time')
plt.xlabel('Order Date')
plt.ylabel('Profit')
plt.show()

# Create a scatter plot with regression line
sns.regplot(data=df, x='Quantity', y='Profit')
plt.title('Profit vs. Quantity')
plt.show()
```

```python
import pandas as pd

# Assuming your original dataframe is named 'df'
# Separate numerical data into one dataframe
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# Separate categorical data into another dataframe
categorical_df = df.select_dtypes(include=['object'])

# Display the numerical dataframe
print("Numerical Data:")
print(numerical_df.head())

# Display the categorical dataframe
print("\nCategorical Data:")
print(categorical_df.head())
```

```python
numerical_df
```

```python
categorical_df
```

```python
# Count NaN values in the numerical dataframe
numerical_nan_count = numerical_df.isna().sum()

# Count NaN values in the categorical dataframe
categorical_nan_count = categorical_df.isna().sum()

# Display the count of NaN values in the numerical dataframe
print("NaN Count in Numerical Data:")
print(numerical_nan_count)

# Display the count of NaN values in the categorical dataframe
print("\nNaN Count in Categorical Data:")
print(categorical_nan_count)
```

```python
# Calculate the mean of each column in the numerical dataframe
numerical_mean = numerical_df.mean()

# Replace NaN values with the mean of each column in the numerical dataframe
numerical_df_filled = numerical_df.fillna(numerical_mean)

# Display the numerical dataframe with NaN values replaced by the mean
print("Numerical Data with NaN Filled:")
print(numerical_df_filled.head())
```

```python
# Count NaN values in the updated numerical dataframe
numerical_nan_count_updated = numerical_df_filled.isna().sum()

# Display the count of NaN values in the updated numerical dataframe
print("Updated NaN Count in Numerical Data:")
print(numerical_nan_count_updated)
```

```python
# Drop the remaining columns with NaN values in the numerical dataframe
numerical_df_filled_dropped = numerical_df_filled.dropna(axis=1)

# Drop rows with NaN values in the numerical dataframe
numerical_df_filled_dropped = numerical_df_filled_dropped.dropna()

# Display the updated numerical dataframe after dropping NaN values
print("Updated Numerical Data after Dropping NaN:")
print(numerical_df_filled_dropped.head())
```

```python
numerical_df_filled_dropped
```

```python


# Create a heatmap of the numerical data
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_df_filled_dropped, cmap='YlGnBu')
plt.title('Heatmap of Numerical Data')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_df_filled_dropped)
plt.title('Heatmap of Numerical Data')
plt.show()
```

```python
# Iterate over each column in the numerical dataframe
for column in numerical_df_filled_dropped.columns:
    # Create a line plot for the current column
    plt.figure(figsize=(8, 6))
    plt.plot(numerical_df_filled_dropped[column])
    plt.xlabel('Index')
    plt.ylabel(column)
    plt.title(f'Line Graph - {column}')
    plt.show()
```

```python
for column in numerical_df_filled_dropped.columns:
    # Create a bar plot for the current column
    plt.figure(figsize=(8, 6))
    plt.bar(numerical_df_filled_dropped.index, numerical_df_filled_dropped[column])
    plt.xlabel('Index')
    plt.ylabel(column)
    plt.title(f'Bar Graph - {column}')
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Iterate over each column in the numerical dataframe
for column in numerical_df_filled_dropped.columns:
    # Create a histogram for the current column
    plt.figure(figsize=(8, 6))
    plt.hist(numerical_df_filled_dropped[column], bins=10)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Histogram - {column}')
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Create box plots for each column
for column in numerical_df_filled_dropped.columns:
    plt.figure(figsize=(8, 6))
    plt.boxplot(numerical_df_filled_dropped[column])
    plt.title(f'Box Plot - {column}')
    plt.show()
```
