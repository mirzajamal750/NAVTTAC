````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 12, 15, 18, 20]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')
plt.show()
````

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 12, 15, 18, 20]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')
plt.show()
````

````python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
````

````python
import matplotlib.pyplot as plt

data = [10, 15, 7, 12, 20, 18, 25, 30, 22, 16]

plt.hist(data, bins=5)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
````

````python
import matplotlib.pyplot as plt

# Count the occurrences of each category in a column
category_counts = df['Category'].value_counts()

# Create a bar chart
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of Categories')
plt.show()
````

````python
import matplotlib.pyplot as plt

# Convert the 'Quantity' and 'Profit' columns to numeric
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')

# Create a scatter plot of the two numerical columns
plt.scatter(df['Quantity'], df['Profit'])
plt.xlabel('Quantity')
plt.ylabel('Profit')
plt.title('Quantity vs Profit')
plt.show()
````

````python
import matplotlib.pyplot as plt

# Convert the date column to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Group the data by month and calculate the total sales
monthly_sales = df.groupby(df['Order Date'].dt.month)['Sales'].sum()

# Create a line plot
plt.plot(monthly_sales.index, monthly_sales.values)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Trend')
plt.show()
````

````python
import seaborn as sns

# Select the numerical columns for the heatmap
numeric_columns = df.select_dtypes(include='number')

# Create a correlation matrix
correlation_matrix = numeric_columns.corr()

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
````
